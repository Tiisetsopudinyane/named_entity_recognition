```python
import fitz  # PyMuPDF
import os

def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    pdf_document.close()
    return text

def merge_pdfs_to_text(pdf_folder, output_txt_path):
    with open(output_txt_path, 'w', encoding='utf-8') as output_file:
        for filename in sorted(os.listdir(pdf_folder)):
            if filename.endswith('.pdf'):
                pdf_path = os.path.join(pdf_folder, filename)
                text = extract_text_from_pdf(pdf_path)
                output_file.write(text + '\n')  # Add a newline between documents

# Example usage
pdf_folder = r'C:\\Users\\JBS LAB\\Downloads\\ScienceDirect_articles_20Aug2024_09-54-50.950'
output_txt_path = './combined_text.txt'
merge_pdfs_to_text(pdf_folder, output_txt_path)
```

```python
import spacy
import pandas as pd
import re

# Load the SpaCy model
nlp = spacy.load('en_core_web_sm')

def process_large_text(text, chunk_size=1000000):
    """Process text in chunks and perform NER."""
    results = []
    for i in range(0, len(text), chunk_size):
        chunk = text[i:i + chunk_size]
        doc = nlp(chunk)
        results.extend([(ent.text, ent.label_) for ent in doc.ents])
    return results

def sanitize_text(text):
    """Remove or replace invalid characters from text."""
    # Replace null bytes and other problematic characters
    text = text.replace('\x00', '')  # Null byte
    text = re.sub(r'[^\x20-\x7E]', '', text)  # Remove non-printable characters
    return text

def read_and_convert_file(txt_file_path):
    """Read text file and convert encoding to UTF-8."""
    with open(txt_file_path, 'r', encoding='latin-1') as file:
        text = file.read()
    # Convert to UTF-8
    return text

def perform_ner_from_file(txt_file_path, excel_file_path):
    # Read and convert text file
    text = read_and_convert_file(txt_file_path)
    
    # Sanitize the text
    text = sanitize_text(text)
    
    # Process the text in chunks
    entities = process_large_text(text)

    # Check if entities is populated correctly
    if not entities:
        print("No entities found. Please check the input text.")
        return
    
    # Create a DataFrame
    df = pd.DataFrame(entities, columns=['Entity', 'Label'])

    # Print DataFrame to verify
    print(df.head())  # Print the first few rows for verification

    # Save to Excel
    df.to_excel(excel_file_path, index=False, engine='openpyxl')

    print(f"NER results have been saved to {excel_file_path}")

# Example usage
txt_file_path = './combined_text.txt'
excel_file_path = './ner_results.xlsx'
perform_ner_from_file(txt_file_path, excel_file_path)
```

```python
import pandas as pd

def search_keywords_in_excel(excel_file_path, keywords):
    """Search for keywords in an Excel file and return matching rows."""
    # Load the Excel file into a DataFrame
    df = pd.read_excel(excel_file_path, engine='openpyxl')

    # Normalize the DataFrame: Convert all text to string and lowercase
    df = df.applymap(lambda x: str(x).lower() if isinstance(x, str) else x)

    # Initialize a list to store matching rows
    matching_rows = []

    # Iterate through each keyword and search in DataFrame
    for keyword in keywords:
        keyword = keyword.lower()
        # Find rows where any column contains the keyword
        mask = df.apply(lambda row: row.astype(str).str.contains(keyword).any(), axis=1)
        matching_rows.append(df[mask])

    # Concatenate all matching rows into a single DataFrame
    if matching_rows:
        result_df = pd.concat(matching_rows).drop_duplicates()
    else:
        result_df = pd.DataFrame()  # Empty DataFrame if no matches

    return result_df

# Example usage
excel_file_path = 'ner_results.xlsx'
keywords = ['cost','employement','infrastructure','service provider']  # List of keywords to search

# Search for keywords and get results
results_df = search_keywords_in_excel(excel_file_path, keywords)

# Print results
if not results_df.empty:
    print("Found matching rows:")
    print(results_df)
else:
    print("No matches found.")
```
